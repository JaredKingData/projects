```python
def generate_cash5_numbers(num_tickets):
    cash5_numbers = []
    for _ in range(num_tickets):
        new_ticket = generate_random_ticket()
        while new_ticket in historical_numbers:
            new_ticket = generate_random_ticket()
        cash5_numbers.append(new_ticket)
        historical_numbers.append(new_ticket)
    return cash5_numbers
```

```python
def generate_random_ticket():
    main_numbers = random.sample(range(1, 36), 5)  # Select 5 main numbers from 1 to 35
    bonus_number = random.randint(1, 5)  # Select a bonus number from 1 to 5
    main_numbers.sort()
    return main_numbers, bonus_number
```

```python
num_tickets = 5
```

```python
tickets = generate_cash5_numbers(num_tickets)
for i, (main_numbers, bonus_number) in enumerate(tickets, 1):
    print(f"Ticket {i}: Main numbers - {main_numbers}, Bonus number - {bonus_number}")
```
